```python
import pandas as ps
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = ps.read_csv('score.csv')
df
```

```python
x = df['이름'].values
y = df['키'].values
```

```python
plt.figure(figsize=(10,5), dpi=50)
plt.plot(x, y, color='skyblue', marker='o', ms=10, mec='red', mfc='yellow', ls='-.')
plt.title('학생별 키')
plt.xlabel('이름')
plt.ylabel('키')
plt.show()
```

```python
kor = df['국어'].values
eng = df['영어'].values
mat = df['수학'].values
sci = df['과학'].values
soc = df['사회'].values
```

```python
import numpy as np
index = np.arange(df.shape[0])
```
